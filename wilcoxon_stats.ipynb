```python
from utils.statistical_test import wilcoxon_statistical_test
import warnings
warnings.filterwarnings('ignore')
import json

def search_vs_nces_stat_tests(kb_emb_model='ConEx'):
    print("### Semantic Bible ###\n")
    with open(f'datasets/semantic_bible/Results/NCES_{kb_emb_model}_Ensemble.json') as file:
        nces = json.load(file)
    f1_nces = nces['SetTransformer_I32+SetTransformer_I64']['f1']['values']
    acc_nces = nces['SetTransformer_I32+SetTransformer_I64']['acc']['values']
    time_nces = nces['SetTransformer_I32+SetTransformer_I64']['time']['values']
    with open('datasets/semantic_bible/Results/CELOE_results.json') as celoe_file:
        celoe = json.load(celoe_file)
    with open('datasets/semantic_bible/Results/EvoLearner_results.json') as evo_file:
        evo = json.load(evo_file)
    f1_celoe = celoe['F-measure']
    acc_evo = evo['Accuracy']
    time_celoe = celoe['Runtime']
    _, p1 = wilcoxon_statistical_test(f1_nces, f1_celoe)
    _, p2 = wilcoxon_statistical_test(time_nces, time_celoe)
    _, p3 = wilcoxon_statistical_test(acc_nces, acc_evo)
    print("\n")
    for metric, p in zip(['F1', 'RunTime', 'Accuracy'], [p1,p2,p3]):
        print(f'On {metric} of Semantic Bible KG, p_value = ', p)
        if p<=0.05:
            print('* Probably different distributions *')
        else:
            print('Probably the same distribution')
        print()
        
    print("### Mutagenesis ###\n")
    with open(f'datasets/mutagenesis/Results/NCES_{kb_emb_model}_Ensemble.json') as file:
        nces = json.load(file)
    f1_nces = nces['SetTransformer_I32+SetTransformer_I128']['f1']['values']
    acc_nces = nces['SetTransformer_I32+SetTransformer_I128']['acc']['values']
    time_nces = nces['SetTransformer_I32+SetTransformer_I128']['time']['values']
    with open('datasets/mutagenesis/Results/EvoLearner_results.json') as evo_file:
        evo = json.load(evo_file)
    f1_evo = evo['F-measure']
    acc_evo = evo['Accuracy']
    time_evo = evo['Runtime']
    _, p1 = wilcoxon_statistical_test(f1_nces, f1_evo)
    _, p2 = wilcoxon_statistical_test(time_nces, time_evo)
    _, p3 = wilcoxon_statistical_test(acc_nces, acc_evo)
    print("\n")
    for metric, p in zip(['F1', 'RunTime', 'Accuracy'], [p1,p2,p3]):
        print(f'On {metric} of Mutagenesis KG, p_value = ', p)
        if p<=0.05:
            print('* Probably different distributions *')
        else:
            print('Probably the same distribution')
        print()
        
    print("### Carcinogenesis ###\n")    
    with open(f'datasets/carcinogenesis/Results/NCES_{kb_emb_model}_Ensemble.json') as file:
        nces = json.load(file)
    f1_nces = nces['SetTransformer_I64+SetTransformer_I128']['f1']['values']
    acc_nces = nces['SetTransformer_I64+SetTransformer_I128']['acc']['values']
    time_nces = nces['SetTransformer_I64+SetTransformer_I128']['time']['values']
    with open('datasets/carcinogenesis/Results/EvoLearner_results.json') as evo_file:
        evo = json.load(evo_file)
    f1_evo = evo['F-measure']
    acc_evo = evo['Accuracy']
    time_evo = evo['Runtime']
    _, p1 = wilcoxon_statistical_test(f1_nces, f1_evo)
    _, p2 = wilcoxon_statistical_test(time_nces, time_evo)
    _, p3 = wilcoxon_statistical_test(acc_nces, acc_evo)
    print("\n")
    for metric, p in zip(['F1', 'RunTime', 'Accuracy'], [p1,p2,p3]):
        print(f'On {metric} of Carcinogenesis KG, p_value = ', p)
        if p<=0.05:
            print('* Probably different distributions *')
        else:
            print('Probably the same distribution')
        print()
        
    print("### Vicodi ###\n")    
    with open(f'datasets/vicodi/Results/NCES_{kb_emb_model}_Ensemble.json') as file:
        nces = json.load(file)
    f1_nces = nces['SetTransformer_I32+SetTransformer_I64+SetTransformer_I128']['f1']['values']
    acc_nces = nces['SetTransformer_I32+SetTransformer_I64+SetTransformer_I128']['acc']['values']
    time_nces = nces['SetTransformer_I32+SetTransformer_I64+SetTransformer_I128']['time']['values']
    with open('datasets/vicodi/Results/EvoLearner_results.json') as evo_file:
        evo = json.load(evo_file)
    f1_evo = evo['F-measure']
    time_evo = evo['Runtime']
    _, p1 = wilcoxon_statistical_test(f1_nces, f1_evo)
    _, p2 = wilcoxon_statistical_test(time_nces, time_evo)
    _, p3 = wilcoxon_statistical_test(acc_nces, acc_evo)
    for metric, p in zip(['F1', 'RunTime', 'Accuracy'], [p1,p2,p3]):
        print(f'On {metric} of Vicodi KG, p_value = ', p)
        if p<=0.05:
            print('* Probably different distributions *')
        else:
            print('Probably the same distribution')
        print()
```
